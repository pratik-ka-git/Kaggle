## *Normalising the data*

```python
def normalize_dataset(data):
    # Select only numerical columns
    numerical_columns = data.select_dtypes(include=[float, int]).columns
    
    # Normalize numerical columns using Min-Max scaler
    scaler = MinMaxScaler()
    data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
    
    # Return the normalized dataset
    return data

# Calling the Function
normalized_data = normalize_dataset(df)
```

```python
# Create a function to identify Outliers
def mark_outliers_zscore(data, column_name, threshold=3.0):
    # Calculate the z-scores for the specified column
    z_scores = (data[column_name] - data[column_name].mean()) / data[column_name].std()

    # Create a new 'Outlier' column based on the z-scores
    data['Outlier'] = abs(z_scores) > threshold

    print(data['Outlier'].unique())
```

# 3. *What was the moving average of the various stocks?*

**The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses.**

```python
ma_days = [50, 120, 200]

for ma in ma_days:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Adj Close'].rolling(ma).mean()
        

fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(15)
fig.set_figwidth(15)

hcltech[['Adj Close', 'MA for 50 days', 'MA for 120 days', 'MA for 200 days']].plot(ax=axes[0,0])
axes[0,0].set_title('hcltech')

sunpharma[['Adj Close', 'MA for 50 days', 'MA for 120 days', 'MA for 200 days']].plot(ax=axes[0,1])
axes[0,1].set_title('sunpharma')

hdfcbank[['Adj Close', 'MA for 50 days', 'MA for 120 days', 'MA for 200 days']].plot(ax=axes[1,0])
axes[1,0].set_title('hdfcbank')

dlf[['Adj Close', 'MA for 50 days', 'MA for 120 days', 'MA for 200 days']].plot(ax=axes[1,1])
axes[1,1].set_title('dlf')

nifty50[['Adj Close', 'MA for 50 days', 'MA for 120 days', 'MA for 200 days']].plot(ax=axes[2,0])
axes[2,0].set_title('nifty50')

fig.tight_layout()
```

**We see in the graph that the best values to measure the moving average are 10 and 20 days because we still capture trends in the data without noise.**

## A] *HCLTECH*

```python
# HCLTECH
plt.figure(figsize=(16,6))
plt.title('Close Price History of HCLTECH')
plt.plot(hcltech['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price INR', fontsize=18)
plt.show()
```

```python
# Create a new dataframe with only the 'Close' column 
data_hcl = hcltech.filter(['Close'])

# Convert the dataframe to a numpy array
dataset_hcl = data_hcl.values

# Get the number of rows to train the model on
hcl_train_data_len = int(np.ceil(len(dataset_hcl) * .80))
hcl_train_data_len
```

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
hcl_scaled_data = scaler.fit_transform(dataset_hcl)

hcl_scaled_data
```

```python
# Create the training data set 
# Create the scaled training data set
hcl_train_data = hcl_scaled_data[0:int(hcl_train_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(200, len(hcl_train_data)):
    x_train.append(hcl_train_data[i-200:i, 0])
    y_train.append(hcl_train_data[i, 0])
    if i<= 201:
        print(x_train)
        print(y_train)
        print()
        
# Convert the x_train and y_train to numpy arrays 
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
# Create the testing data set
# Create a new array containing scaled values
hcl_test_data = hcl_scaled_data[hcl_train_data_len - 200: , :]
# Create the data sets x_test and y_test
x_test = []
y_test = dataset_hcl[hcl_train_data_len:, :]
for i in range(200, len(hcl_test_data)):
    x_test.append(hcl_test_data[i-200:i, 0])
    
# Convert the data to a numpy array
x_test = np.array(x_test)

# Reshape the data
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))

# Get the models predicted price values 
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
# Plot the data
train = data_hcl[:hcl_train_data_len]
valid = data_hcl[hcl_train_data_len:]
valid['Predictions'] = predictions

# Visualize the data
plt.figure(figsize=(16,6))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
